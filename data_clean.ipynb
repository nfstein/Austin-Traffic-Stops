```python
import pandas as pd
```

```python
df = pd.read_csv('2016_Racial_Profiling_Dataset_Citations.csv')
df.columns
```

```python
df = df[['CITATION NUMBER', 'OFF FROM DATE', 'OFF TIME', 'REASON FOR STOP', 'VL STREET NAME']]
```

```python
df.head(10)
```

```python
df = df.set_index('CITATION NUMBER')
df.count()
```

```python
df = df.dropna(subset=['VL STREET NAME'])
```

```python
grouped_df = df.groupby('VL STREET NAME').count().reset_index()
```

```python
# filters out addresses without numerical address attached (erroneously includes others like "EAST 7TH STREET")
# keeps the main corner indicators / and 'AT' which might not have numbers
# could filter out '7th' with some form of split that looks for a numerical word that does not contain 'th'
good_addresses = []
bad_addresses = []
for address in grouped_df['VL STREET NAME']:
    if any(char.isdigit() for char in address):  
        good_addresses.append(address)
    # entries at corners might not contain numbers
    elif '/' in address:  
        good_addresses.append(address)
    elif ' AT' in address and address[-2:] != 'AT':   #many have AT at the end of the entry with no second road listed
        good_addresses.append(address)
    else: bad_addresses.append(address)
bad_addresses
```

```python
df = df.rename(columns = {'VL STREET NAME': 'ADDRESS'})
```

```python
df = df[df['ADDRESS'].isin(good_addresses)]
df.head()
```

```python
grouped_df = grouped_df[grouped_df['VL STREET NAME'].isin(good_addresses)]
grouped_df = grouped_df[['VL STREET NAME', 'OFF FROM DATE']]
grouped_df = grouped_df.rename(columns = {'VL STREET NAME': 'ADDRESS', 'OFF FROM DATE': 'STOPS'})
grouped_df = grouped_df.sort_values('STOPS', ascending=False)#[2:] #first 2 results are ambiguous
```

```python
maxStops = grouped_df['STOPS'].max()
def To_Weight(x):
    return x/maxStops
grouped_df['WEIGHT'] = grouped_df['STOPS'].apply(To_Weight)
```
